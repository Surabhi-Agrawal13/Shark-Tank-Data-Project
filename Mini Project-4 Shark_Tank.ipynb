### Q-2. Max, min , mean of asked amount,asked equity, asked valuation? 

```python
Value1 = ST_data[["pitcher_ask_amount"]].agg([max,min,"mean"])
print("The Minimum, Maximum and Mean value of Asked amount is", Value1)
```

```python
Value2 = ST_data[["ask_equity"]].agg([max,min,"mean"])
print("The Minimum, Maximum and Mean value of Asked Equity is", Value2)
```

```python
Value3 = ST_data[["ask_valuation"]].agg([max,min,"mean"])
print("The Minimum, Maximum and Mean value of Asked Valuation is", Value3)
```
